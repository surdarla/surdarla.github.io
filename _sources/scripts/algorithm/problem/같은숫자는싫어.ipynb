## Python

````python
# input setting
problem_num = "ê°™ì€ìˆ«ìëŠ”ì‹«ì–´"
import os, sys

path = os.getcwd() + f"\\txt\\{problem_num}" + ".txt"
sys.stdin = open(path, "r")
arr = list(map(int, sys.stdin.readline().strip().split(",")))
````

````python
def solution(arr):
    answer = []
    for i in arr:
        if len(answer) == 0:
            answer.append(i)
        elif answer[-1] != i:
            answer.append(i)
        else:
            continue
    return answer
````

````python
solution(arr)
````
